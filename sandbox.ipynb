```python
import os
import json
from youtube_script import YoutubeManager
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())

yt_manager = YoutubeManager()

```

```python
all_videos = {}
all_playlists = yt_manager.playlist_list(os.getenv("USER_ID"))
all_playlists
```

```python
for playlist in all_playlists:
    playlist_id = playlist["id"]
    playlist_vids = yt_manager.playlist_elements(playlist_id)
    all_videos[playlist_id] = playlist_vids
```

```python
with open("FULLALL_tracks.json", "w") as f:
    json.dump(all_videos, f, indent=4)
```

```python
import pandas as pd

def pandas_load(all_videos_resp) -> pd.DataFrame:
    df = pd.DataFrame()
    for playlist_id, playlist_vids in all_videos_resp.items():
        batch = pd.DataFrame.from_dict(playlist_vids)
        batch["playlist_id"] = playlist_id
        df = pd.concat([df, batch])
    return df

def pandas_process(df: pd.DataFrame) -> pd.DataFrame:
    df = df.drop(columns=["kind"])
    
    return df

df = pandas_load(all_videos)
df = pandas_process(df)
df
```
